```python
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
from collections import Counter
from nltk.corpus import brown
import os
from sklearn.preprocessing import normalize
nltk.download('brown')
nltk.download('stopwords')
nltk.download('punkt')
```

```python
topics = os.listdir('./20news-18828/')

dataset = []
for topic in topics:
  for files in os.walk("./20news-18828/"+topic):
    for file in files[2]:
      dataset.append([topic,file])
```

```python
data = pd.DataFrame(dataset)
data.columns = ['topic','docID']
data.head()
```

```python
words = []
for file in dataset:
  topic = file[0]
  idDoc = file[1]
  with open('./20news-18828/'+topic+'/'+idDoc, 'r', encoding="utf8", errors='ignore') as f:
    text = f.read()
    #[word for line in f for word in line.split()]
    
    #removing numbers
    text = ''.join(c for c in text if not c.isdigit())
    file_content = ''.join(c for c in text if c != '_')
    
    #tokenization removing punctuation
    #tokens = word_tokenize(file_content)
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(file_content)
    
    #removing stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if not w in stop_words]
    
    tokens = [w for w in tokens if len(w) > 2]
    
    #steamming
    porter = PorterStemmer()
    stems = []
    for t in tokens:    
      stems.append(porter.stem(t))
    
    doc_words = set(stems)
    counts = Counter(stems)
    
    for w in doc_words: 
      words.append([topic,idDoc,w,counts[w]])
```

```python
w_data = pd.DataFrame(words)
w_data.columns = ['topic','docID','word','count']
print(len(w_data))
w_data.head()
```

```python
red_data = w_data.ix[:90000, :]

pivot_data = pd.pivot_table(red_data, values = 'count', index = ['docID'], columns = ['word'], fill_value = 0)

pivot_data.head()
```

```python
def TF_IDF(test_data):
    #Total number of docs:
    test_no_docs = len(test_data)
    IDF_dict = {name: np.log(float(test_no_docs) / (1 + len(test_data[test_data[name] > 0]))) for name in test_data.columns}
    new_data = pd.DataFrame()
    for name in test_data.columns:
        new_data[name] = test_data[name] * IDF_dict[name]
    return(new_data)

TF_IDF_data = TF_IDF(pivot_data)
TF_IDF_data.head()
```

```python
norm_data = normalize(TF_IDF_data)
norm_data = pd.DataFrame(norm_data)
norm_data.columns = TF_IDF_data.columns
norm_data.head()
```

```python
from sklearn import cluster

kmeans = cluster.KMeans(n_clusters = 20
                       , init = 'k-means++'
                       , n_init = 10
                       , tol = 0.0001
                       , n_jobs = -1
                       , random_state = 1)

kmeans.fit(norm_data)
```

```python
centroids = pd.DataFrame(kmeans.cluster_centers_)
centroids.columns = TF_IDF_data.columns
centroids
```

```python
for i in range(0, len(centroids)):
    print(centroids.ix[i,:].sort_values(ascending = False)[0:10])
    print("\n")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline

for i in range(0, len(centroids.ix[0])):
    wordcloud = WordCloud(background_color = 'white')
    kMeansWordCloud = wordcloud.generate_from_frequencies(centroids.ix[i])
    plt.figure()
    plt.imshow(kMeansWordCloud)
    plt.axis("off")
```
