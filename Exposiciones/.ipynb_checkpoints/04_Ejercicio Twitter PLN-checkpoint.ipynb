## Definici√≥n de Funciones

```python
def ObtenerTweets(palabra,numero_de_Tweets,lenguaje):
    #Se define las listas que capturan la popularidad
    popularidad_list = []
    numeros_list = []
    lista_tweets = []
    numero = 1
    for tweet in tweepy.Cursor(api.search, palabra, lang=lenguaje).items(numero_de_Tweets):
        try:
            lista_tweets.append(tweet._json)
            with open('tweet_json_PLN.txt', 'w') as file:
                file.write(json.dumps(lista_tweets, indent=4))
            
            # Se toma el texto, se hace el analisis de sentimiento
            # y se agrega el resultado a las listas
            
            analisis = TextBlob(tweet.text)
            analisis = analisis.sentiment
            
            # Retorna una tupla de la forma (polarity, subjectivity ) donde polarity 
            # es un float en el rango [-1.0, 1.0] y subjectivity es un float 
            # en el rango [0.0, 1.0] donde 0.0 es muy objetivo 
            # y 1.0 es muy subjetivo.
            
            polaridad = analisis.polarity
            popularidad_list.append(polaridad)
            numeros_list.append(numero)
            numero = numero + 1

        except tweepy.TweepError as e:
            print(e.reason)

        except StopIteration:
            break
    return (numeros_list, popularidad_list, numero)
```

```python
def VerDatos():
  lista = []
  with open('tweet_json_PLN.txt', encoding='utf-8') as json_file:  
      all_data = json.load(json_file)
      for each_dictionary in all_data:
          tweet_id = each_dictionary['id']
          text = each_dictionary['text']
          favorite_count = each_dictionary['favorite_count']
          retweet_count = each_dictionary['retweet_count']
          created_at = each_dictionary['created_at']
          user = "@" + each_dictionary['user']['screen_name']
          lista.append({'tweet_id': str(tweet_id),
                               'user': str(user),
                               'text': str(text),
                               'favorite_count': int(favorite_count),
                               'retweet_count': int(retweet_count),
                               'created_at': created_at,
                              })
          tweet_json = pd.DataFrame(lista, columns = 
                                    ['tweet_id', 'user', 'text', 
                                     'favorite_count', 'retweet_count', 
                                     'created_at'])
  return tweet_json
  
```

```python
def GraficarDatos(numeros_list,popularidad_list,numero):
    axes = plt.gca()
    axes.set_ylim([-1, 2])
    
    plt.scatter(numeros_list, popularidad_list)
    popularidadPromedio = (sum(popularidad_list))/(len(popularidad_list))
    popularidadPromedio = popularidadPromedio
    time  = datetime.now().strftime("A : %H:%M\n El: %m-%d-%y")
    plt.text(0, 1.25, 
             "Sentimiento promedio:  " + str(popularidadPromedio) + "\n" + time, 
             fontsize=12, 
             bbox = dict(facecolor='none', 
                         edgecolor='black', 
                         boxstyle='square, pad = 1'))
    
    plt.title("Sentimientos sobre " + palabra + " en Twitter")
    plt.xlabel("Numero de Tweets")
    plt.ylabel("Sentimiento")
    plt.show()
    
```
