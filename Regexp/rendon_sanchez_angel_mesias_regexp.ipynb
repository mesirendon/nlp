## Expresiones regulares

```python
# Librerias a usar
import pandas as pd
import re
```

```python
# Lectura del archivo
data = pd.read_csv("twitter.tsv", delimiter = "\t", quoting = 3)
del data["Unnamed: 0"]

# 5 primeros registros...
data.head()
```

```python
# Utilidad para imprimir la lista a partir de una expresion regular
def printer(regexp):
    count = 0;
    print("Indice :", "Texto")
    print("=================\n")
    for row in data.itertuples():
        if(regexp.search(row.text)):
            count += 1
            print(row.Index, ":", row.text)

    print("Coincidencias:", count)
```

Encontrar las líneas de los tweets que comienzan con letras mayúsculas y que tengan un usuario de twitter o un hashtag al final.

```python
regexp = re.compile(r'^[A-Z].+([@|#]\S+)$')
```

Iterar sobre el set de datos aplicando la expresion regular a cada text e imprimiendo solo aquellos que concuerdan.

```python
printer(regexp)
```

Usando phyton encontrar líneas de tweets que tengan números telefónicos

```python
# Numeros telefonicos con indicativo formato colombiano
regexp = re.compile(r'\s(\(?\+(\d{1,2}[\s|-]?)?\d{1,3}\)?[\s|-]?)?(\d{3}[\s|-]?)?\d{3}[\s|-]?\d{4}')
```

```python
printer(regexp)
```

Encontrar todos los tweets que tienen emoticones (ayuda: Usar unicode block: x1F000 - x2F000)

```python
regexp = re.compile(r'[\u263a-\U0001f645]')
```

```python
printer(regexp)
```

## Tokenization

Implementar una función considerando los siguientes casos:

·         Dividir tokens por espacios, “tab”, y “enters”

§  ' this is a tweet ' -> 'this', 'is', 'a', 'tweet'

·         Dividir por emoticones:

§  haha 😂😂! -> 'haha', '😂😂', '!'

·         NO divider hashtags

§  #don'tSplit

·         Separar signos de puntuación del texto (tokesn individuales):

§  'hey!' -> 'hey', '!'

·         Separar “Contractions” usando el “apostrophe”:

§  "that's fine" -> 'that', "'s", 'fine'

§  "don't do that" -> 'do', "n't", 'do', 'that'

```python
def splitter(text):
    tokens = re.split(r'[\s|\t|\n|\u263a-\U0001f645]+', text)
    return tokens
```

```python
print("Indice :", "Tokens")
print("=================\n")
for row in data.itertuples():
    print(row.Index, ":", row.text, splitter(row.text))
```
